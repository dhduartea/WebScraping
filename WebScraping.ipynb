- Importar las librerías en tu script de Python:

```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
```

- Utiliza la librería requests para enviar una solicitud HTTP a la página web y obtener el contenido HTML:

```python
url = "https://en.wikipedia.org/wiki/Web_scraping"
response = requests.get(url)
html_content = response.content
```

```python
print(html_content)
```

- Crea un objeto BeautifulSoup pasándole el contenido HTML obtenido y el parser que deseas utilizar (por ejemplo, "html.parser"):

```python
soup= BeautifulSoup(html_content, 'html.parser')
```

```python
print(soup.prettify)
```

- Utiliza selectores CSS o XPath para seleccionar los elementos que deseas extraer. Puedes utilizar los métodos find() o find_all() de Beautiful Soup para esto. Aquí hay algunos ejemplos:

```python
# Ejemplo utilizando un selector CSS
titles = soup.select('.titulo')  # selecciona todos los elementos con la clase 'titulo'
print(titles)
```

```python
# quiero guardar el código css en una variable 
css_selector = '.titulo'
titles = soup.select(css_selector)
print(titles)



```

```python

# Ejemplo utilizando XPath
links = soup.find_all('a')  # selecciona todos los elementos 'a'
print(links)
```
